```python
#dropping columns 
df.drop(columns=[ 'Maximum sustained wind speed',
                 'Direction of the longest radius of 50kt winds or greater',
                'The longeast radius of 50kt winds or greater',
                'The shortest radius of 50kt winds or greater', 
                 'Direction of the longest radius of 30kt winds or greater',
                'The longeast radius of 30kt winds or greater',
               'The shortest radius of 30kt winds or greater',
                 'Indicator of landfall or passage'], axis=1, inplace=True)  
df
```

- We are combining 'year','month','day' and 'hour' to a single column date representing the date.


- datatype "datetime64ns" helps to perform meaningful time-based analysis, visualization, and comparisons.

- To access to the original columns we can use 'df.DATE.dt.year', df.DATE.dt.month, df.DATE.dt.day and df.DATE.dt.hour.

```python
# since year,month,day,hour  is integer data type convert into a single column 'DATE'(datatype datetime64)
df['DATE']=pd.to_datetime(df[['year','month','day','hour']])   
df.drop(['year','month','day','hour'],axis=1,inplace=True)       #dropping the columns 
df
```

- we have to change the index position of DATE column after the international number id   for that  we can perform insert function 

```python
new_df=df.pop('DATE')
df.insert(2, 'DATE',new_df)
df
```

```python
#renaming the columns
df.rename(columns={'Latitude of the center':'HORIZONTAL LATITUDE','Longitude of the center':'VERTICAL LONGITUDE'},inplace=True)
df.rename(columns={'International number ID':'ID','grade':'GRADE','Central pressure':'CP'},inplace=True)
df
```

```python
#renaming and dropping the first unnamed column since we alredy have rows number as an index  
df.rename(columns={'Unnamed: 0':'index'},inplace=True)
df.drop('index',axis=1,inplace=True)
df
```

```python
# Define the mapping between GRADE and intensity
intensity= {
    'Low': 0,
    'Moderate': 1,
    'High': 2
}

# lets create a Function to calculate intensity based on CP and GRADE
def Grade_intensity(row):
    grade = row['GRADE']
    return intensity.get(grade, -1)  # -1 for unknown intensity

# Apply  to the DataFrame
df['Intensity_Encoded'] = df.apply(Grade_intensity, axis=1)
df.head(10)
```

- The above code convert the categorical 'GRADE' datatype to numerical value

- Then we check the intensity mapping using 'Grade_intensity' function."-1" value indicates that the 'GRADE value doesnt found in the mapping

```python
plt.scatter(df['CP'], df['GRADE'], c=df['Intensity_Encoded'], cmap='viridis')
plt.xlabel('CENTRAL PRESSURE ')
plt.ylabel('GRADE')
plt.title('INTENSITY BASED ON CP AND GRADE ')
plt.colorbar(label='Intensity')
plt.show()
```

- Here we have plotted  (scatter plot) the intensity based on 'CP' and 'GRADE each point is colored based on the calculated intensity  of encoded values 
    - "-1" value suggests that the 'GRADE' values in the  dataset  that are not accounted for in the predefined mapping.
